小結 : group 與 filt 的差異在於 group 可以從全體到指定項 filt 則需要指定條件項

```python
filt = public_df['Country'] == 'India'
public_df.loc[filt]['LanguageWorkedWith'].str.contains('Python')
```

```python
public_df.loc[filt]['LanguageWorkedWith'].str.contains('Python').sum() # sum 用以統計 True 的數量
```

```python
country_grp.get_group('India')['LanguageWorkedWith'] # 嘗試利用 group 達到相同效果 (使用前面已經做好的 country_grp)
```

```python
country_grp.get_group('India')['LanguageWorkedWith'].str.contains('Python').sum() # 驗證可以達到相同效果
```

```python
country_grp['LanguageWorkedWith'].str.contains('Python').sum() # 出現 SeriesGroupBy 的錯誤是因為 group 沒有經過 get_group 無法轉換成 Dataframe
```

```python
country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum()) # 改以 apply 寫一個簡短的副程式 一行一行送進去給 sum() 處理便可迴避上面的錯誤達到效果
```

```python
country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum()).loc['India'] # 成功先統計全體再進行特定搜尋
```

練習 : What % of people from each country know Python?

```python
country_respondents = public_df['Country'].value_counts() # 首先知道所有有國籍且記錄在 Data 裡的人作國家總人數統計
country_respondents
```

```python
country_uses = country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum()) # 知道每個國家使用 Python 的人數
country_uses
```

```python
python_df = pd.concat([country_respondents, country_uses], sort=False) # 利用 concat 結合兩個 Dataframe 但是需要特別注意的是預設 axis 為 'rows'會造成後面的資料直接嫁接在第一個資料後面 所以需要改成 'columns' 因為兩個 Dataframe 資料性質不相關
python_df
```

```python
python_df = pd.concat([country_respondents, country_uses], axis='columns', sort=False) # 利用 concat 結合兩個 Dataframe 但是需要特別注意的是預設 axis 為 'rows'會造成後面的資料直接嫁接在第一個資料後面 所以需要改成 'columns' 因為兩個 Dataframe 資料性質不相關
python_df
```

```python
python_df.rename(columns = {'Country': 'NumRespondents', 'LanguageWorkedWith': 'NumKnowsPython'}, inplace=True)
python_df
```

```python
python_df['PctKnowsPython'] = (python_df['NumKnowsPython']/python_df['NumRespondents'])*100 # 直接將列索引之間所有元素進行對應計算
python_df
```

```python
python_df.sort_values(by = 'PctKnowsPython', ascending=False, inplace=True) # ascending=False 會讓數值大的排在前面 ascending=True 會讓數值小的排在前面
python_df.head(50)
```
