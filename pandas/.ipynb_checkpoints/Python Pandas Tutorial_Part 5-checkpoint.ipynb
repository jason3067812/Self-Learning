## Updating Rows and Columns - Modifying Data Within DataFrames

```python
import pandas as pd
import numpy as np
```

以下為 Python dictionary 各種寫法

```python
person = {"first": "Corey",
          "last": "Schafer",
          "email": "CoreySchafer@gmail.com"}
```

```python
people = {"first": ["Corey"],
          "last": ["Schafer"],
          "email": ["CoreySchafer@gmail.com"]}
```

```python
people = {"first_name": ["Corey", "Jane", "John"],
          "last_name": ["Schafer", "Doe", "Doe"],
          "email": ["CoreySchafer@gmail.com", "JaneDoe@gmail.com", "JohnDoe@gmail.com"]}
```

```python
df = pd.DataFrame(people) # pandas 可以直接讀取 dictionary 格式
```

```python
df
```

```python
df.columns # 取得列索引
```

```python
df.columns = [x.upper() for x in df.columns] # 將列索引一一指定給 x 並改成大寫 再存回原本的 df.columns
df
```

```python
df.columns = df.columns.str.replace('_', ' ') # 將所有列索引字串部分做搜尋'_'並取代為' '
df
```

```python
# 改回原資料 如上的反操作
df.columns = [x.lower() for x in df.columns]
df.columns = df.columns.str.replace(' ', '_')
df
```

```python
df.rename(columns = {'first_name':'first', 'last_name':'last'}) # 用 rename 更改列索引名稱(注意括號)且這裡不會改到原資料
```

```python
df # 如上述上面的操作並未改到原資料
```

```python
df.rename(columns = {'first_name':'first', 'last_name':'last'}, inplace=True) # 如 pandas 很多函式的操作加入 inplace=True 才可改到原資料
df
```

```python
df.loc[2] = ['John', 'Smith', 'JohnSmith@gmail.com'] # loc 最直觀的更改資料的方法 注意 diemension 須和原資料該範圍的 diemension 相同
df
```

```python
df.loc[2, ['last', 'email']] = ['Doe', 'JohnDoe@gmail.com'] # 操作概念同上
df
```

```python
df.loc[2, 'last'] = 'Smith' # 操作概念同上
df
```

```python
df.at[2, 'last'] = 'Doe' # at 處理效果與 loc 相同 (選自己習慣用的)
df
```

```python
filt = (df['email'] == 'JohnDoe@gmail.com') # 如同之前設計一個資料過濾器
df[filt]['last'] = 'Smith' # 試著直接以陣列指定方式寫入新資料發現無法寫入(pandas不可使用這種方法)
```

```python
df # 雖然上面是警告 但是確實對於原資料並沒有實際的效果
```

```python
filt = (df['email'] == 'JohnDoe@gmail.com') # 如同之前設計一個資料過濾器
df.loc[filt, 'last'] = 'Smith' # 改以 loc 修正
```

```python
df
```

```python
df['email'].str.lower() # 可以搜尋列索引所有字串並轉成小寫
```

```python
df # 上面的操作並沒有改到原資料
```

```python
df['email'] = df['email'].str.lower() # 此操作才會改到原資料
df
```

```python
df['email'].apply(len) # apply 可以執行函式(自己寫的也可以) 概念是需指定列索引後將一項一項給函式輸入
```

```python
df.apply(len) # 這裡不指定列索引 會回傳所有列索引的長度
```

```python
df.apply(len, axis='rows') # 指定處理軸為 rows 效果同上回傳所有列索引的長度
```

```python
df.apply(len, axis='columns') # 指定處理軸為 columns 回傳所有行索引的長度
```

```python
def update_email(email): # 自己設計處理函式(單項輸入)
    return email.upper()
```

```python
df['email'] = df['email'].apply(update_email) # 不須給輸入 apply 會自動給
df
```

```python
df['email'] = df['email'].apply(lambda x: x.lower()) # lambda 為 python 的簡易版 def 適合寫極短的運算式
df
```

```python
df.apply(pd.Series.min) # apply 取用 pandas 標準處理函式 (pd.Series.min: 查找列索引中字串長度最短的 若長度相同則已先找到的作為輸出)
```

```python
df.apply(lambda x: x.min()) # 同樣也可以用 lambda 寫到相同效果(這裡 min 用法只有在 pandas 中可以這樣用)
```

```python
df.applymap(len) # 利用 applymap 可以將函式效果應用於整個 Dataframe
```

```python
df.applymap(str.lower) # 概念同上 但是並不會改到原資料
```

```python
df
```

```python
df['first'].map({'Corey': 'Chris', 'Jane': 'Mary'}) # map 只能用於處理列索引的元素 diemension 不一定要等於原資料或一定要能在原資料中找到 但是沒有被處理的元素將以 NaN 呈現
```

```python
df['first'].replace({'Corey': 'Chris', 'Jane': 'Mary'}) # 用法與上面相似 但是但是沒有被處理的元素將以原資料的形式呈現 (同樣不管是上面或者現在的操作還不會改到原資料)
```

```python
df['first'] = df['first'].replace({'Corey': 'Chris', 'Jane': 'Mary'}) # 要這樣寫才會改到原資料
df
```
