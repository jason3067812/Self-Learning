```python
def to2sCompStr(num, bitWidth):
    num &= (2 << bitWidth-1)-1 # mask
    formatStr = '{:0'+str(bitWidth)+'b}'
    ret =  formatStr.format(int(num))
    return ret


def decode_data(decimal):
    
    
    data = to2sCompStr(decimal, 32)
#     print("data:",data)
#     print("length of data:",len(data))
#     print(" ")

    data1 = data[0:8]
    data2 = data[8:16]
    data3 = data[16:24]
    data4 = data[24:32]
    #print("before:",data1,data2,data3,data4)

    data1 = int(data1,2)
    data2 = int(data2,2)
    data3 = int(data3,2)
    data4 = int(data4,2)
    #print("after:",data1,data2,data3,data4)
    
    return data1, data2, data3, data4
```

```python
import openpyxl
import numpy
import cv2
import numpy as np
from PIL import ImageFont, Image, ImageDraw


# 開啟工作簿
print("loading excel......")
wb = openpyxl.load_workbook(r"C:\Users\jason\OneDrive\桌面\coordinates.xlsx")

# 獲取表單
sh = wb['coordinates']


print("最大行數:",sh.max_row)
print("最大列數:",sh.max_column)
print("########################################################################################################################")

rows = sh.rows


# average_temp=[]
# max_temp=[]
# min_temp=[]
count=0
for row in list(rows):                # 遍歷每行資料

    count=count+1
    if count>1:
        actually_img = count-1
        case = []                     # 用於存放一行資料
        for c in row[3:]:
            data = c.value
            
            #print(data)
            
            new_data = decode_data(data)
            
            case.append(new_data[3])
            case.append(new_data[2])
            case.append(new_data[1])
            case.append(new_data[0])
            
        
        print(f"Data_{actually_img}")
        print(case)

        img_name = f"D:\\sleep_detection_project\\PIR Camera\\csv files\\new_img\\img_{actually_img}.jpg"
        width = 24
        height = 32
        img = np.zeros((width, height))

        # 遍历每个像素点，并进行赋值
        cc=0
        i=0
        j=0
        for i in range(width):
            for j in range(height):

                img[i][j] = case[cc]
                cc=cc+1

        cv2.imwrite(img_name, img)
        print("writing into an image....")
        print(" ")
```
