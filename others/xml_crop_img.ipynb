```python
import os
import sys
import xml.etree.ElementTree as ET
import glob
import cv2


def xml_crop_img(indir,img_dir,new_path):
    
    os.chdir(indir)
    annotations = os.listdir('.')
    annotations = glob.glob('*.xml')
    
    for i, file in enumerate(annotations):
        
        print(file)
        head = os.path.splitext(file)
        img_path_name=img_dir + '\\' + head[0]+".bmp"
        if os.path.exists(img_path_name):
            img_path_name=img_path_name
        else:
            img_path_name=img_dir + '\\' + head[0]+".jpg"
            
        image = cv2.imread(img_path_name)
                            
        # actual parsing
        in_file = open(file, encoding="utf-8")
        tree=ET.parse(in_file)
        root = tree.getroot()
        filename = root.find('filename').text  
        n=0
        for obj in root.iter('object'):
            n=n+1
            
            current = list()
            name = obj.find('name').text  
            xmlbox = obj.find('bndbox')
            xn = xmlbox.find('xmin').text   
            xx = xmlbox.find('xmax').text
            yn = xmlbox.find('ymin').text
            yx = xmlbox.find('ymax').text
                    
            # 裁切圖片
            crop_img = image[int(yn): int(yx), int(xn): int(xx)]

            crop_path=new_path+"\\"+f"{name}"
            if not os.path.isdir(crop_path):
                os.makedirs(crop_path)
                
            cv2.imwrite(crop_path + "/" + f"{head[0]}_{n}.jpg", crop_img)
                          
                
indir=r'D:\OCR\Steel_coil\test750_xml'   #xml目錄
img_dir=r"D:\OCR\Steel_coil\test750_img" #img目錄
new_path=r"D:\OCR\Steel_coil\crop_id\test"  #存放crop圖片的位置


xml_crop_img(indir,img_dir,new_path)
```
